## Статистический анализ данных

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**нулевая гипотеза:** средние пользовательские рейтинги XOne и PC равны

**альтернативная гипотеза:** средние пользовательские рейтинги XOne и PC НЕ равны

```python
XOne = data_updated.query('platform == "XOne"')['user_score'].dropna()
XOne.mean()
```

```python
PC = data_updated.query('platform == "PC"')['user_score'].dropna()
PC.mean()
```

```python
import numpy as np
print(np.var(XOne))
print(np.var(PC))
```

разница дисперсий около 44%, поэтому дисперсии считаем неравными, equal_var = False

```python
from scipy import stats as st

alpha = .05 #возьмем в качестве параметра статзначимости общепринятое пороговое значение 5%

results = st.ttest_ind(XOne, PC, equal_var = False) #используем ind, т.к. это сравнение среднего двух генеральных совокупностей по независимым выборкам из них 

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

минимальный уровень статзначимости, на котором нулевая гипотеза отвергается, равен 0,54, что больше, чем статзначимость, следовательно, средние пользовательских рейтингов XOne и PC однозначны равны

#### Средние пользовательские рейтинги жанров Action и Sports разные

**нулевая гипотеза:** средние пользовательские рейтинги жанров Action и Sports равны

**альтернативная гипотеза:** средние пользовательские рейтинги жанров Action и Sports НЕ равны

```python
action = data_updated.query('genre == "action"')['user_score'].dropna()
action.mean()
```

```python
sports = data_updated.query('genre == "sports"')['user_score'].dropna()
sports.mean()
```

```python
print(np.var(action))
print(np.var(sports))
```

разница дисперсий около 16%, поэтому дисперсии считаем равными, equal_var = True

```python
alpha = .05

results = st.ttest_ind(action, sports, equal_var = True)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

минимальный уровень статзначимости, на котором нулевая гипотеза отвергается, равен 3.71, что значительно меньше, чем статзначимость, следовательно, средние пользовательские рейтинги жанров Action и Sports не равны

**2. средние пользовательские рейтинги жанров Action и Sports равны**

- нулевая гипотеза **противоречит** данным (уровень значимости меньше статистической значимости, 3.71 < 0.05)

Нулевая гипотеза была сформулирована так, чтобы поставить между значениями (XOne и PC; action и sports) знак равенства.
